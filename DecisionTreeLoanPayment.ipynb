```python

import warnings 
warnings.filterwarnings("ignore")
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

data = pd.read_csv("E:\DSTUTORIALS\loanpayment.csv")
#print(data.head(5))

data['Principal'].unique()
data['loan_status'].unique()

le=preprocessing.LabelEncoder()
data['loan_status']=le.fit_transform(data['loan_status'])
data['Gender']=le.fit_transform(data['Gender'])
data['education']=le.fit_transform(data['education'])
data['past_due_days']=le.fit_transform(data['past_due_days'])
print(data)

#print(data.head(5))
data.groupby('loan_status')['Gender'].agg(['count'])

data2 = data
data2.drop('Loan_ID', axis=1, inplace=True)

data2.drop('effective_date', axis=1, inplace=True)
data2.drop('due_date', axis=1, inplace=True)
data2.drop('paid_off_time', axis=1, inplace=True)

y = data2['loan_status'].values
X = data2.drop('loan_status', axis=1).values

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2,
                                                    random_state = 0)
dt = tree.DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred = dt.predict(x_test)
dt_score_train = dt.score(x_train, y_train)
print("Training score: ",dt_score_train)
dt_score_test = dt.score(x_test, y_test)
print("Testing score: ",dt_score_test)
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))


#Random Forest Classifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
rfc_score_train = rfc.score(x_train, y_train)
print("Training score: ",rfc_score_train)
rfc_score_test = rfc.score(x_test, y_test)
print("Testing score: ",rfc_score_test)

scores = pd.DataFrame({
        'Model'          : ['Decision Tree', 'Random Forest'],
        'Training_Score' : [dt_score_train, rfc_score_train],
        'Testing_Score'  : [dt_score_test, rfc_score_test]
    })
scores.sort_values(by='Testing_Score', ascending=False)








```
