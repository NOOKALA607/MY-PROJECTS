```python
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
```

```python
algorithms=[]
algorithms.append(("LogisticRegression:",LogisticRegression()))
algorithms.append(("DecisionTree:",DecisionTreeClassifier()))
algorithms.append(("RandomForest:",RandomForestClassifier()))
algorithms.append(("KNeighbrs:",KNeighborsClassifier()))
algorithms.append(("Naive_bayes:",GaussianNB()))
algorithms.append(("Support_Vector_Machine:",SVC()))
```

```python
algorithms
```

```python
import warnings
warnings.filterwarnings("ignore")
for name,model in algorithms:
    kf=KFold(n_splits=10)
    result=cross_val_score(model,x_train,y_train,scoring="accuracy",cv=kf)
    print(name,(result.mean())*100)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipe_logistic=Pipeline([("StandardScaler",StandardScaler()),
                       ("LogisticRegression:",LogisticRegression())])
pipe_decision=Pipeline([("StandardScaler",StandardScaler()),
                        ("DecisionTree:",DecisionTreeClassifier())])
pipe_random=Pipeline([("StandardScaler",StandardScaler()),
                     ("RandomForest:",RandomForestClassifier())])
pipe_neighbors=Pipeline([("StandardScaler",StandardScaler()),
                         ("KNeighbors:",KNeighborsClassifier())])
pipe_naive=Pipeline([("StandardScaler",StandardScaler()),
                    ("naive_bayes",GaussianNB())])
pipe_svm=Pipeline([("StandardScaler",StandardScaler()),
                   ("svm",SVC())])
pipeline=[pipe_logistic,
          pipe_decision,
          pipe_random,
          pipe_neighbors,
          pipe_naive,
          pipe_svm]
pipeline
```

```python
pipe_dict=dict({0:"LogisticRegression",
               1:"DecisionTree",
               2:"RandomForest",
               3:"KNeighbors",
               4:"Naive_bayes",
               5:"Support_vector_machine"})
for i,model in enumerate(pipeline):
    model.fit(x_train,y_train)
    
    print("{} accuracy : {}".format(pipe_dict[i],model.score(x_train,y_train)))
```

```python
best_classifier=""
best_accuracy=0
for i,model in enumerate(pipeline):
    if model.score(x_test,y_test)>best_accuracy:
        best_accuracy=model.score(x_test,y_test)
        best_classifier=i
print("The best classifier is {}".format(pipe_dict[best_classifier]))
```

```python
model=RandomForestClassifier(n_estimators=100,
                             random_state=100,
                            max_features=7,
                             criterion="gini")
model.fit(x_train,y_train)
y_train_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)
from sklearn.metrics import accuracy_score
train_acc=accuracy_score(y_train,y_train_pred)
test_acc=accuracy_score(y_test,y_test_pred)
print("training_accuracy:",train_acc)
print("testing_accuracy:",test_acc)
print("error_in_train",100-int((train_acc)*100))
print("error_in_test",100-int((test_acc)*100))
```
