```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data1['department']=le.fit_transform(data1['department'])
data1['region_no']=le.fit_transform(data1['region_no'])
data1['education']=le.fit_transform(data1['education'])
data1['gender']=le.fit_transform(data1['gender'])
data1['recruitment_channel']=le.fit_transform(data1['recruitment_channel'])

data1
```

```python
data=data1.iloc[:,0:12]
target=data1.iloc[:,12]
data
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,target,train_size=0.7)

```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import KFold,cross_val_score
```

```python
all_alg=[]
lr=LogisticRegression()
dt=DecisionTreeClassifier()
rf=RandomForestClassifier()
svc=SVC()
knn=KNeighborsClassifier()
all_alg.append(("LogisticRegression:",lr))
all_alg.append(("DecisionTreeClassifier:",dt))
all_alg.append(("RandomForestClassifier:",rf))
all_alg.append(("SVC:",svc))
all_alg.append(("KNN:",knn))
all_alg
```

```python

import warnings
warnings.filterwarnings("ignore")
for name,model in all_alg:
    kf=KFold(n_splits=10)
    cv_res=cross_val_score(model,x_train,y_train,scoring="accuracy",cv=kf)
    
    print(name,cv_res.mean())
```
