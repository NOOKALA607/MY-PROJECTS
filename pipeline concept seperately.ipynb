```python
#PipeLine is used to automate the machine learning process. 
#we can find the best algorithm with better accuracy

import pandas as pd
df=pd.read_csv("C:\\Users\\DELL\\JUPYTER FILES\\MY_PROJECTS\\HR_analytics_with_feats_selection.csv")
```

```python
x=df.drop(["Unnamed: 0"],axis=1)
x
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
x["region_no"]=le.fit_transform(x["region_no"])
x
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
pipe_logistic=Pipeline([("standascaler",StandardScaler()),
              ("Logistic",LogisticRegression())])

pipe_decision=Pipeline([("standascaler",StandardScaler()),
              ("decision",DecisionTreeClassifier())])

pipe_random=Pipeline([("standascaler",StandardScaler()),
              ("randomforest",RandomForestClassifier())])

pipe_svm=Pipeline([("standascaler",StandardScaler()),
              ("svm",SVC())])

pipe_knn=Pipeline([("standascaler",StandardScaler()),
              ("knn",KNeighborsClassifier())])

pipe_naive=Pipeline([("standascaler",StandardScaler()),
              ("naive",GaussianNB())])

pipelines=[pipe_logistic,pipe_decision,pipe_random,pipe_svm,pipe_knn,pipe_naive]
```

```python
pipedict=dict({0:"Logistic",1:"Decision",2:"randomforest",3:"svm"
              ,4:"knn",5:"naive"})
data=x.iloc[:,:7]
y=x.iloc[:,7]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(data,y,test_size=0.7)
for i,model in enumerate(pipelines):
    model.fit(x_train,y_train)
    model.predict(x_test)
    print("{} accuracy: {}".format(pipedict[i],model.score(x_train,y_train)))
```

```python
best_accuracy=0
best_classifier=0
best_pipeline=""
for i,model in enumerate(pipelines):
    
    if model.score(x_test,y_test)>best_accuracy:
        best_accuracy=model.score(x_test,y_test)
        best_pipeline=model
        best_classifier=i
print("Best Classifier is {}".format(pipedict[best_classifier]))
```
